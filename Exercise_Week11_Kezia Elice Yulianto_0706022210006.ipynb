## Load Data

We will use the dataset of the transaction in a certain store. You can get the dataset here:
https://gist.githubusercontent.com/Harsh-Git-Hub/2979ec48043928ad9033d8469928e751/raw/72de943e040b8bd0d087624b154d41b2ba9d9b60/retail_dataset.csv

```python
# load the data set and show the first five transaction
df = pd.read_csv('https://gist.githubusercontent.com/Harsh-Git-Hub/2979ec48043928ad9033d8469928e751/raw/72de943e040b8bd0d087624b154d41b2ba9d9b60/retail_dataset.csv')
df.head()
```

# Get the set of product that has been purchased


```python
purchased_products = set(item for row in df.values for item in row)
print(purchased_products)
```

## Preprocess Data

In this step, we will transform our dataset so that we will have a one hot encoding based on the purchased products.

```python
# create an itemset based on the products
itemset = set(purchased_products)

# encoding the feature
itemset_encoded = []
for row in df.values:
    row_encoded = {item: 1 if item in row else 0 for item in itemset}
    itemset_encoded.append(row_encoded)

itemset_encoded[0]
```

```python
# create new dataframe from the encoded features
df_new = pd.DataFrame(itemset_encoded)

# show the new dataframe
df_new.head()
```

```python
# Since, the encoded dataframe consist of the empty column. We will drop the NaN column or u can use the index.
df_new.drop(columns=[float('nan')], inplace=True)
df_new.head()
```

Since, the encoded dataframe consist of the empty column. We will drop the NaN column or select all columns other than the first column.

## Apriori Algorithm

We will use appriori algorithm to determine the frequently purchased products.
For this case study, we will min_support=0.2

```python
#Set threshold value untuk digunakan dalam penghitungan support
from mlxtend.frequent_patterns import apriori, association_rules

min_support = 0.2
frequent_itemsets = apriori(df_new, min_support=min_support, use_colnames=True)
frequent_itemsets
```

Then we will generate association rule of the frequent itemset based on confidence level with the threshold=0.6

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6)
rules.drop(columns=['zhangs_metric'], inplace=True, errors='ignore')
rules
```

1. Antecedent Support: This tells you how often the item (or combination of items) on the left side of the rule (antecedent) appears in the dataset.
Example (Milk  Cheese): If the antecedent is Milk, a support of 0.501587 means that Milk appears in about 50.16% of the transactions.
2. Consequent Support: This shows how often the item (or combination of items) on the right side of the rule (consequent) appears in the dataset.
Example (Milk  Cheese): For Cheese, the support of 0.501587 means that Cheese is purchased in 50.16% of all transactions.
3. Support (Rule Support): This measures how often the combination of antecedent and consequent appear together in the dataset.
Example (Milk  Cheese): The support of 0.304762 means that Milk and Cheese together are bought in 30.48% of the transactions.
4. Confidence: Confidence is a measure of the reliability or strength of an association rule. It quantifies the likelihood that if itemset X is purchased, itemset Y will also be purchased. A high confidence value indicates that itemset Y is often purchased when itemset X is purchased, suggesting a strong association or dependency between the two. On the other hand, a low confidence value suggests a weaker association.
Example (Milk  Cheese): A confidence of 0.607595 means that when Milk is bought, theres about 60.76% chance that Cheese will be bought too.
5. Lift: Lift is a metric that measures the strength of association between two itemsets (A and B) while taking into account the expected or random occurrence of B given A. It is used to determine if the association between A and B is significant beyond what would be expected by chance. Example (Milk  Cheese): A lift of 1.211344 means that Milk and Cheese are 21.13% more likely to be bought together than expected by chance. Lift greater than 1 indicates that the presence of A has a positive effect on the likelihood of B being purchased, suggesting a stronger association between A and B than would be expected by chance. Lift equal to 1 means that there is no association between A and B beyond what would be expected by chance. Lift less than 1 suggests that the presence of A has a negative effect on the likelihood of B being purchased, indicating a weaker association between A and B.
6. Leverage: Leverage measures the difference between the observed frequency of co-occurrence of two items (X and Y) in transactions and the frequency that would be expected if X and Y were independent of each other. In other words, it quantifies how much the occurrence of X and Y together deviates from what would be expected by chance. Example (Milk  Cheese): A leverage of 0.053172 means that Milk and Cheese appear together 5.32% more often than we would expect by chance. A positive leverage value indicates that the occurrence of X and Y together is more frequent than what would be expected by chance, suggesting a positive association. A leverage value of 0 suggests that the co-occurrence of X and Y is as expected under independence, meaning there is no association. A negative leverage value indicates that the occurrence of X and Y together is less frequent than expected, suggesting a negative association or avoidance.
7. Conviction: Conviction measures the ratio of the expected frequency that X occurs without Y to the observed frequency that X occurs without Y. In essence, it quantifies how much the presence of X implies the absence of Y. X high conviction value indicates that the rule X => Y is highly dependent, and A rarely occurs without B. Example (Milk  Cheese): A conviction of 1.270148 means that the rule Milk  Cheese is 1.27 times more likely to be true than if they were independent. A conviction value close to 1 suggests that there is no association between X and Y, as the presence of X does not affect the presence or absence of Y. A high conviction value (> 1) indicates that the presence of X is strongly dependent on the presence of Y, implying a strong association. A low conviction value (< 1) suggests that the presence of X is not strongly dependent on the presence of Y, implying a weaker association.

These metrics help assess the strength and significance of relationships between items, such as Milk  Cheese, in transactional data. Higher values in confidence, lift, and conviction indicate stronger associations between items.
