```python
# g
df_avg_fighters = df_mlbb[df_mlbb['role'] == 'fighter']['physical_atk'].mean()
df_avg_tanks = df_mlbb[df_mlbb['role'] == 'tank']['physical_atk'].mean()
print("Average physical attack for Fighters: " + str(df_avg_fighters))
print("Average physical attack for Tanks: " + str(df_avg_tanks))

if df_avg_fighters > df_avg_tanks:
    print("Fighters have higher average physical attack than Tanks.")
elif df_avg_fighters < df_avg_tanks:
    print("Tanks have higher average physical attack than Fighters.")
```

7. I have a character with this specification:
defense_overall 6.0<br>
offense_overall 7.0<br>
skill_effect_overall 4.0<br>
difficulty_overall 3.0<br>
movement_spd 250<br>
magic_defense 9<br>
mana 380<br>
hp_regen 37<br>
physical_atk 119<br>
physical_defense 24<br>
hp 2689<br>
attack_speed 0.75<br>
mana_regen 16.0<br>

Please predict its role using DecisionTree classification algorithm <br>

```python
# 7

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Encode the 'role' column
encoder = LabelEncoder()
df_mlbb['role'] = encoder.fit_transform(df_mlbb['role'])

# Separate features and target variable
X = df_mlbb.drop(['hero_name', 'role', 'win_rate',	'pick_rate',	'ban_rate', 'release_year'], axis=1)
y = df_mlbb['role']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)

# Create a Decision Tree Classifier
dt_classifier = DecisionTreeClassifier()

# Train the model
dt_classifier.fit(X_train, y_train)

# Make predictions on the test set
y_pred = dt_classifier.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: ", accuracy)

# Predict the role for the given character
new_character = [[6.0, 7.0, 4.0, 3.0, 250, 9, 380, 37, 119, 24, 2689, 0.75, 16.0]]
predicted_role = dt_classifier.predict(new_character)

# Get the predicted role (dicoding)
predicted_role_label = encoder.inverse_transform(predicted_role)
print("Predicted role for the character: ", predicted_role_label[0])
```
